## matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
y1 = y
y2 = np.cos(x)
```

```python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,linestyle='--',label='cos')
plt.legend()
plt.show()
```
