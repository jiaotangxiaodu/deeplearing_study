## 阶跃函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    return np.array(x>0,dtype='int')
```

```python
step_function(np.array([-1,1,2]))
```

```python
x = np.arange(-2,2,0.1)
y = step_function(x)
plt.plot(x,y)
plt.show()
```

## sigmoid函数

```python
def sigmoid(x):
    return 1./(1+np.exp(-x))
```

```python
sigmoid(np.array([-1,1,2]))
```

```python
x = np.arange(-10,10,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
```

## ReLU函数

```python
def relu(x):
    return np.maximum(0,x)
```

```python
relu(np.array([-1,1,2]))
```

```python
x = np.arange(-10,10,0.1)
y = relu(x)
plt.plot(x,y)
plt.show()
```
